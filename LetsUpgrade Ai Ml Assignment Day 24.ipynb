# Titanic Decision Tree


```python
import pandas as pd
from sklearn import tree
import numpy as np
```

```python
dataset = pd.read_csv("C:/Users/0703s/Downloads/Day 24 AI ML/train.csv")
dataset.head()
```

```python
dataset1=dataset.drop(['PassengerId', 'Pclass','Name','SibSp','Parch','Ticket','Cabin','Embarked'], axis = 1)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
dataset1['Sex'] = le.fit_transform(dataset1['Sex'])
```

```python
dataset1.head()
```

```python
predictor=pd.DataFrame([dataset1['Sex'], dataset1['Age'], dataset1['Fare']]).T
model=tree.DecisionTreeClassifier(max_depth=5)
```

```python
model.fit(X=predictor,y=dataset1['Survived'])
```

```python
with open("Treefile1.dot",'w')as f:
    f= tree.export_graphviz(model, feature_names=['Sex','Age','Fare'],out_file=f);
```

```python
model.score(X=predictor,y=dataset1['Survived'])
```

# Bank personal loan descision tree

```python
dataset=pd.read_excel("C:/Users/0703s/Downloads/Day 24 AI ML/Bank_Personal_Loan_Modelling.xlsx", sheet_name='Data')
dataset.head()
```

```python
dataset1=dataset.drop(['ID', 'ZIP Code'], axis = 1)
dataset1.head()
```

```python
predictor=pd.DataFrame([dataset1['Experience'], dataset1['Age'], dataset1['Income'],dataset1['Family'], dataset1['CCAvg'], dataset1['Education'],dataset1['Mortgage'], dataset1['Securities Account'], dataset1['CD Account'],dataset1['Online'], dataset1['CreditCard']]).T
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_model=RandomForestClassifier(n_estimators=1000,max_features=2,oob_score=True)
features=['Experience','Age','Income','Family','CCAvg','Education','Mortgage','Securities Account','CD Account','Online','CreditCard']
rf_model.fit(X=predictor,y=dataset1['Personal Loan'])
```

```python
print('The oob score is..........',rf_model.oob_score_)
```

```python
for feature,imp in zip(features,rf_model.feature_importances_):
    print(feature,imp)
```

```python
rf_model.fit(X=predictor,y=dataset1['Personal Loan'])
```

```python
model=tree.DecisionTreeClassifier(max_depth=5)
model.fit(X=predictor,y=dataset1['Personal Loan'])
```

```python
with open("TreefileBankLoan.dot",'w')as f:
    f= tree.export_graphviz(model, feature_names=['Experience','Age','Income','Family','CCAvg','Education','Mortgage','Securities Account','CD Account','Online','CreditCard'],out_file=f);
```

```python
model.score(X=predictor,y=dataset1['Personal Loan'])
```
